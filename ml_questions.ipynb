In this problem I am using K means clustering algorithm to classify wheather a website is suspecious or not.


```python
import pandas as pd
import numpy as np
df=pd.read_csv('phishing_data.csv')
df

```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2)
predicted=km.fit_predict(df)
```

```python
predicted
```

If values in predicted array are zero then we will say that website is phishing website.
Result variable will be 1 if website is phishing and zero if website is safe

```python
Result=[]
for i in range(0,2456):
    if predicted[i]==0:
        Result.append(1)
    else:
        Result.append(0)
    
```

```python
count=0
for i in range(0,2456):
    if Result[i]==df['Result'][i]:
        count =count+1
```

```python
df['Result']=Result
df.head()
```

```python
#accuracy of the model
accuracy=(count/2455)*100
accuracy
```
